```python
#1 Programs on selection and Iteration operations.

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)


def count_digits(num):
    count = 0
    while num > 0:
        num //= 10
        count += 1
    return count


def is_palindrome(num):
    num_str = str(num)
    reversed_str = num_str[::-1]
    return num_str == reversed_str


number = int(input("Enter an integer: "))

if number % 2 == 1:
    fact = factorial(number)
    digit_count = count_digits(fact)
    print("Factorial of", number, "is:", fact)
    print("Number of digits in the factorial:", digit_count)
else:
    if is_palindrome(number):
        print(number, "is a palindrome!")
    else:
        print(number, "is not a palindrome!")



```

```python
#4 Given two strings, PRINT (YES or NO) whether the second string can be obtained from the first by deletion of none, one or more characters.

def can_obtain_by_deletion(first_string, second_string):
    i = 0
    j = 0

    while i < len(first_string) and j < len(second_string):
        if first_string[i] == second_string[j]:
            j += 1
        i += 1

    if j == len(second_string):
        return "YES"
    else:
        return "NO"

first_string = "vishal"
second_string = "ish"
result = can_obtain_by_deletion(first_string, second_string)
print(result)
```

```python
#7 Program to check if the given list is in Ascending order or Not.

def is_ascending(lst):
    n = len(lst)
    for i in range(1, n):
        if lst[i] < lst[i-1]:
            return False
    return True

numbers = [1, 2, 3, 4, 5]
if is_ascending(numbers):
    print("The list is in ascending order.")
else:
    print("The list is not in ascending order.")
```

```python
#9 Python program to convert a tuple to a string.

def tuple_to_string(tuple_value):
    string_value = ''.join(str(element) for element in tuple_value)
    return string_value

#example
my_tuple = ('H', 'e', 'l', 'l', 'o')
result = tuple_to_string(my_tuple)
print(result)
```

```python
#10 Python program to reverse a tuple.

def reverse_tuple(tuple_value):
    reversed_tuple = tuple(reversed(tuple_value))
    return reversed_tuple

# Example
my_tuple = (1, 2, 3, 4, 5)
reversed_tuple = reverse_tuple(my_tuple)
print(reversed_tuple)

```

```python
#12 Python program to check if a set is a subset of another set.

def is_subset(set1, set2):
    return set1.issubset(set2)

# Example
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}
if is_subset(set1, set2):
    print("set1 is a subset of set2.")
else:
    print("set1 is not a subset of set2.")
```

```python
#13  Dictionaries and its operations

#1.Creating a Dictionary:
my_dict = {'name': 'John', 'age': 25, 'country': 'USA'}
another_dict = dict(name='Alice', age=30)

#2.Accessing Values:
my_dict = {'name': 'John', 'age': 25, 'country': 'USA'}
print(my_dict['name'])
print(my_dict.get('age'))

#3.Updating Values:
my_dict = {'name': 'John', 'age': 25, 'country': 'USA'}
my_dict['age'] = 30
print(my_dict)

#4.Adding New Key-Value Pairs:
my_dict = {'name': 'John', 'age': 25}
my_dict['country'] = 'USA'
print(my_dict)

#5.Removing Key-Value Pairs:
my_dict = {'name': 'John', 'age': 25, 'country': 'USA'}
del my_dict['age']
print(my_dict)

my_dict.pop('country')
print(my_dict)

#6.Checking Key Existence:
my_dict = {'name': 'John', 'age': 25}
print('name' in my_dict)
print('country' in my_dict)

#7.Dictionary Size:
my_dict = {'name': 'John', 'age': 25, 'country': 'USA'}
print(len(my_dict))

#8.Iterating over a Dictionary:
my_dict = {'name': 'John', 'age': 25, 'country': 'USA'}

# Iterate over keys
for key in my_dict:
    print(key)


# Iterate over values
for value in my_dict.values():
    print(value)


# Iterate over key-value pairs
for key, value in my_dict.items():
    print(key, value)

```

```python
#14  Python program to iterate over dictionaries using for loops.

my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}

# Iterate over the keys
for key in my_dict:
    print(key)

# Iterate over the values
for value in my_dict.values():
    print(value)

# Iterate over both keys and values
for key, value in my_dict.items():
    print(key, value)
```
